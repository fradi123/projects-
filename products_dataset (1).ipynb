2. Drop Unnecessary columns that we will not use during our analysis

```python
# Drop columns :
products_data = products_data.drop(columns=[
    'total_urgency_count', 'urgency_text_rate', 'badges_count',
    'has_urgency_banner', 'urgency_text', 'badge_local_product',
    'badge_product_quality', 'badge_fast_shipping', 'uses_ad_boosts',
    'mean_product_rating_count', 'mean_discount', 'merchant_rating_count',
    'theme', 'crawl_month','shipping_is_express', 'merchant_profile_picture',
    'product_variation_size_id','merchant_rating'
], errors='ignore')
```

```python
products_data.head()
```

3. Rename "inventory_total" to "inventory_total_start" and "tags" to "additional_information"

```python
products_data.rename(columns={'inventory_total': 'inventory_total_start', 'tags':'additional_information'}, inplace=True)
```

4. Update "inventory_total_start" to be greater than "units_sold"

```python
if 'inventory_total_start' in products_data.columns and 'units_sold' in products_data.columns:
    products_data['inventory_total_start'] = products_data['units_sold'] + abs(products_data['units_sold'].median())
```

5. Compute ratings percentages, create new columns : "ratings_over_three" and "ratings_below_three", and drop the old columns. This will make our dataset easier to analyze and will make it more concise.

```python
# Create the new columns
for col in ['rating_one_count', 'rating_two_count', 'rating_four_count', 'rating_five_count']:
    if col in products_data.columns and 'rating_count' in products_data.columns:
        products_data[col] = (products_data[col] / products_data['rating_count']) * 100

if all(col in products_data.columns for col in ['rating_four_count', 'rating_five_count', 'rating_one_count', 'rating_two_count']):
    products_data['ratings_over_three'] = products_data['rating_four_count'] + products_data['rating_five_count']
    products_data['ratings_below_three'] = products_data['rating_one_count'] + products_data['rating_two_count']
```

```python
# Drop columns 'rating_x_count' since we won't need them anymore :
products_data = products_data.drop(columns=[
    'rating_three_count', 'rating_four_count', 'rating_five_count',
    'rating_one_count', 'rating_two_count', 'rating_count'
], errors='ignore')
```

6. Identify missing values in each column, display them and handle them one by one

```python
# Identify missing values in each column
missing_values = products_data.isnull().sum()
```

```python
# Display columns with missing values
missing_columns = missing_values[missing_values > 0]
missing_columns_details = missing_columns.to_frame(name='MissingCount').reset_index()
missing_columns_details.rename(columns={'index': 'Column'}, inplace=True)

missing_columns
missing_columns_details
```

```python
# Handle missing values column by column
missing_handling_strategies = {
    'product_color': 'Unknown',
    'product_variation_size_id': 'Unknown',
    'origin_country': 'Unknown',
    'merchant_name': 'Unknown',
    'merchant_info_subtitle': 'No info',
    'merchant_profile_picture': 'No picture',
    'ratings_over_three': products_data['ratings_over_three'].median(),
    'ratings_below_three': products_data['ratings_below_three'].median()
}
```

```python
# Apply the handling strategies
for column, value in missing_handling_strategies.items():
    if column in products_data.columns:
        products_data.fillna({column:value}, inplace=True)
```

```python
# Recheck for missing values in all columns after handling
final_missing_values = products_data.isnull().sum()
final_missing_values[final_missing_values > 0]
```

9. Add the "inventory_total_end" column (amounts equal to "inventory_total_start" - "units_sold" for each row)

```python
#Checking for rows in 'merchant_info_subtitle' missing the '%' symbol and replace them with 'No feedback available'. Replace NaN with empty string for safe checks.
products_data['merchant_info_subtitle'] = products_data['merchant_info_subtitle'].fillna('')
missing_percent = ~products_data['merchant_info_subtitle'].str.contains('%', na=False)
products_data.loc[missing_percent, 'merchant_info_subtitle'] = 'No feedback available'
```
