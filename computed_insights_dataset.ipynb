7.Dropping uncessary columns

8.Make all values of averagediscount positive 

```python
def transform_to_positive(df, column_name):

    if column_name in df.columns:
        df[column_name] = df[column_name].abs()  # Apply absolute value
    else:
        print(f"Column '{column_name}' not found in dataframe!")
    return df

# Apply the transformation to the "averagediscount" column
computed_insights = transform_to_positive(computed_insights, "averagediscount")
```
